 # גרף ראשון 
צרו מידע על פי ההנחיות הבאות ולאחר מכן צרו גרף קווי.

(Time)ציר ה x יהיה בטווח 0-3 בקפיצות של 0.02

(Voltage)ציר ה y יהיה חישוב: sin(2 * pie * Time) + 1


הוסיפו כותרות ורשת(grid)

```python
#import
import numpy as np
import matplotlib.pyplot as plt

#data
time = np.arange(0.0, 3.0,0.02)
voltage = 1 + np.sin(2 * np.pi * time)

#creating figure and subplots
fig, axs = plt.subplots(1)

#style (suptitle is A style that belongs to the figure)
fig.suptitle('Voltage Usage')
axs.plot(time,voltage)
axs.grid()

plt.show()
```

### בונוס:

הוסיפו קו שיתאר את ההיסטוגרמה בצורה הטובה ביותר
צרו את הקו בסגנון +-, למדו איך עושים זאת בעזרת ספריית scipy.


```python
import scipy.stats as stats
data = np.zeros(100)
for i in range(0,99):
    data[i + 1] = np.random.randint(data[i], data[i] + 4)

density = stats.gaussian_kde(data)
fig, ax = plt.subplots(1,1)
n, x, _ = ax.hist(data,histtype = 'step',density = True)
ax.plot(x, density(x), linestyle = '--')
plt.style.use('ggplot')
plt.show()
```

בכל משבצת במפה הציגו את הערך בצבע כחול, הוסיפו את האות F בכל משבצת.

```python
a = np.random.random((8, 8)) #random data

plt.figure(figsize=(10,10))
plt.rcParams["axes.grid"] = False
ax = plt.gca()
im = ax.imshow(a, cmap='hot')

#set yticks and xticks to 1-8
ax.set_xticklabels(np.arange(0,9,1))
ax.set_yticklabels(np.arange(0,9,1))
ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)

#styles
plt.title("Heat map",fontsize = 30)
ax.yaxis.set_label_position("right")
plt.ylabel("Values",fontsize = 25)

#create axis cell value and text displayed
for i in range(8):
    for j in range(8):
        text = ax.text(j, i,'%.2fF' % a[i, j],
                       ha="center", va="center", color="blue")
plt.show()
```

## גרף חמישי

צרו מידע רנדומלי על טמפרטורות במהלך השנה(לפי חודשים)

המידע יראה כך:

הטמפרטורה הממוצעת היא לא הממוצע של הטמפרטורה המקסימלית והמינימלית באותו חודש.

הציגו את המידע עבור שנה אחת בלבד

עליכם להציג את המידע בשתי צורות:
    
דרך ראשונה - על ידי שלושה קווים בצבעים שונים

ערכי ציר ה y יהיו 0-32 בקפיצות של 2

הוסיפו רשת(grid) רק בציר ה y(לרוחב)

דרך שנייה - צרו טבלה מתחת לגרף שתציג את המידע כולו(כולל כותרות בצירים)

הוסיפו כותרות בשני הדרכים

```python
#A way to solve with subplots

max_temp = []
min_temp = []
avg_temp = []
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

for i in range(12):
    ma = np.random.randint(15, 32)
    max_temp.append(ma)
    mi = np.random.randint(5, ma - 1)
    min_temp.append(mi)
    av = np.random.randint(mi + 1, ma)
    avg_temp.append(av)

fig, ax =plt.subplots(nrows = 2,ncols = 1, figsize=(8,8))

data=[max_temp,min_temp,avg_temp]
column_labels=months
ax[1].axis('off')
ax[1].table(cellText=data,colLabels=column_labels,loc="center",rowLabels=['max','min','avg'])
ax[1].set_title('Table')



ax[0].plot(months, max_temp, label = 'Max Temp', color = 'crimson')
ax[0].plot(months, min_temp, label = 'Min Temp', color = 'dodgerblue')
ax[0].plot(months, avg_temp, label = 'Avg Temp', color = 'darkorange')
ax[0].set_title('Weather')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Temperature')
ax[0].grid(axis = 'x')
ax[0].legend(loc = 3, fontsize = 8)
ax[0].set_yticks(range(0, 32, 2))



plt.tight_layout()
plt.show()
```

### דרך לפתור ללא subplots:

```python
#A way to solve without subplots
plt.figure(figsize=(15,10))
columns =  ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
rows = ["Maximum Temperature", "Average Temperature", "Minimum Temperature"]
values = np.arange(0, 32, 2)

temp = -32*np.round(np.random.random_sample((12, 3)),3) + 32
min_tmp = np.amin(temp, axis=1)
max_tmp = np.amax(temp, axis=1)
avg_tmp = np.median(temp, axis = 1)


data = [max_tmp,avg_tmp,min_tmp]
colors = ["r","g","b"]
index = np.arange(len(columns)) + 0.3
bar_width = 0.4
n_rows = len(data)
for row in range(n_rows):
    plt.plot(index, data[row], bar_width, color=colors[row])
plt.table(cellText=data, rowLabels=rows, rowColours=colors,
          colLabels=columns, loc='bottom')
plt.subplots_adjust(left=0.25, bottom=0.25)
plt.ylabel("Temperatures in Celzuis")
plt.yticks(values)
plt.xticks([])
plt.title('Teperatures - plot and table')
plt.grid(True,color='black')
plt.show()
```

### שלב ראשון
צרו מערכי x ו - y שההפרש בין כל אחד מהערכים שלהם שווה ואחיד לאורך המערך - סדרה חשבונית. שני המערכים צריכים להיות זהים באורכם, אך לא בהכרח זהים בערכיהם.  


### שלב שני
בחלק זה נשתמש בפעולה מתוך ספריית numpy הנקראת np.meshgrid: פונקציה זו לוקחת שני מערכים X ו- Y ויוצרת שני מערכים חדשים: 
 - מערך X יהיה מערך דו ממדי בו כל מערך פנימי מכיל את הסדרה החשבונית שיצרתם תחילה במערך x. כמות המערכים הפנימיים תהיה שווה לכמות הערכים במערך x הראשוני. 
 
 דוגמה: אם מערך x הוא [1,2,3] אז מערך X יהיה [[1,2,3],[1,2,3],[1,2,3]]
 
 - מערך Y יהיה גם הוא מערך דו מימדי שמכיל מערכים פנימיים כמספר הערכים היחידים במערך y, אך בכל מערך יש העתקים של כל אחד מהערכים, כל אחד במערך נפרד.
 
 דוגמה: אם מערך y הוא [1,2,3] אז מערך Y יהיה [[3,3,3],[2,2,2],[1,1,1]].
 
### שלב שלישי
צרו את מערך z כתוצאה: X^2 + Y^2. (רמז: למערכים בספריית numpy יכולת לבצע פעולות המשלבות את כל המערך בפעולות מתמטיות ללא לולאות).

 
 ### שלב רביעי
חקרו וצרו את הגרף contour3D בעזרת שלושת המערכים שיצרתם.

עצבו את הגרף בגודל 15 יחידות לאורך ו15 יחידות לרוחב, יכולים לבחור כותרות וצבעים משלכם.

```python
#creating the values distribution:
x_values = np.linspace(-6, 6, 30)
y_values = np.linspace(-5, 5, 30)

#Creating meshgrid
X, Y = np.meshgrid(x_values, y_values)

#creating Z array
Z = X**2 + Y**2

#extracting figures and axes
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='hot')

#style
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
