# plt.plot

## Line chart

צרו linechart (רמז - קראו על הפונקציה plt.plot) בעל המאפיינים הבאים:

- צבע הקו יהיה ירוק

- רוחב הקו יהיה 5

- סגנון הקו יהיה dashdot

בסוף נציג את הגרף עם הפונקציה plt.show() (קראו עליה באינטרנט)

```python
plt.plot(df.Day,df.Temperature, color='green',linewidth=2, linestyle='dashed')
plt.xlabel('Day')
plt.ylabel('Temperature', rotation=0, labelpad=40)
plt.title('Weather')
plt.show()
```

## תרגול

צרו linechart בעל המאפיינים הבאים:

- הצבע יהיה ירוק

- המרקר יהיה "+"

- סגנון הקו יהיה dashdot

כדי שיהיה לנו יותר קל, נשים את העמודות במשתנים. 

```python
x = df.Day
y = df.Temperature

plt.plot(x, y, color='green', marker='+', linestyle='dashdot')
plt.show()
```

## Format String

קראו בקישור הבא על Format string - דרך נוספת לעצב את מאפייני הגרף שלנו אך הפעם נשתמש בstring פשוט במקום מספר פרמטרים שונים: https://python-course.eu/numerical-programming/formatting-plot-in-matplotlib.php

```python
x = df.Day
y = df.Temperature

plt.plot(x, y, '-.+g') # הסדר לא משנה, כי הוא יכול לזהות כל רצף.
plt.show()
```

שנו את גודל המרקר ל20

```python
plt.plot(x, y, color='green', marker='+', linestyle='dashdot', markersize=20)
plt.show()
```

הוסיפו לגרף עמעום (opacity) של 0.5

```python
plt.plot(x, y, color='green', marker='+', linestyle='dashdot', markersize=20, alpha=0.5)
plt.show()
```

## Axes labels, Legend, Grid

### legend - מתאר לגרף

ניצור קצת מידע...

```python
df = pd.DataFrame({'Day': ['SUN','MON','TUE','WED','THU','FRI','SAT'],
                  'Max_Temperature': [50,51,52,48,47,49,46],
                  'Min_Temperature': [43,42,40,44,33,35,37],
                  'Avg_Temperature': [45,48,48,46,40,42,41]})

max_t = df.Max_Temperature
min_t = df.Min_Temperature
avg_t = df.Avg_Temperature
days = df.Day

df
```

צרו גרף שמציג 3 קווים

ציר הX יהיה days

ציר הY יהיה טמפרטורה

הוסיפו כותרות מתאימות לצירים

הוסיפו כותרת weather לגרף

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t)
plt.plot(days, min_t)
plt.plot(days,avg_t)
plt.show()
```

הוסיפו לכל אחד מהקווים תווית מתאימה שתתאר אותו.

הוסיפו לגרף מתאר (באננגלית - legend)

מקמו את המתאר במקום הטוב ביותר (רמז - loc)

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
# We use the label property
plt.plot(days,max_t, label="Max") 
plt.plot(days, min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
# And the legent function
plt.legend(loc='best') # We use the property loc for locating the legend, the default is 'best'
plt.show()
```

כעת נכנס עמוק יותר לתוך הפונקציה plt.legend

כנסו לדוקומנטציה באינטרנט ווסיפו את הפרמטרים הבאים:

צל

גודל כתב

חפשו מאפיינים נוספים לבחירתכם :)

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
# We use the label property
plt.plot(days,max_t, label="Max") 
plt.plot(days, min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
# And the legent function
plt.legend(loc='best', shadow=True, fontsize='small') 
plt.show()
```

### Grid - רקע רשת

הוסיפו grid לגרף שלכם

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

# We use the label property
plt.plot(days,max_t, label="Max") 
plt.plot(days, min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

# And the legent function
plt.legend(loc='best', shadow=True, fontsize='small') 

# And the grid function
plt.grid()

plt.show()
```

## Bar Chart

ניצור מידע

```python
df = pd.DataFrame({'Company': ['GOOGLE', 'AMAZON', 'MSFT', 'FACEBOOK'],
                  'Revenue': [90, 136, 89, 27],
                  'Profit': [40,2,34,12]})

company = df.Company
revenue = df.Revenue
profit = df.Profit

df
```

כעת ניצור גרף עמודות מהמידע.

בנתיים ציר הX שלנו ייצג את החברות, בהמשך נוכל לשנות את זה

```python
plt.bar(company, revenue, label='Revuenue')
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.legend()
plt.show()
```

מה אם נרצה להוסיף את ההכנסות לגרף שלנו?

```python
plt.bar(company, revenue, label='Revuenue')
plt.bar(company, profit, label='Profit')
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.legend()
plt.show()
```

שנו את הגרף כך שעמודות הרווח וההכנסות יהיו מופרדות אחת מהשנייה (רמז השתמשו במערך של numpy)

```python
import numpy as np
xpos = np.arange(len(company))
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit ') # lower the width and moving each of them aside
plt.xticks(xpos, company) # Replaces the titles
plt.legend()
plt.show()
```

ואם נרצה שהגרף שלנו יהיה אופקי?

```python
ypos = np.arange(len(company))
plt.ylabel('revenue(bln)')
plt.title('US Tech Stocks')
plt.barh(ypos-0.2, revenue, height=0.4, label='Revenue')
plt.barh(ypos+0.2, profit, height=0.4, label='Profit ') # lower the height and moving each of them aside
plt.yticks(ypos, company) # Replaces the titles
plt.legend()
plt.show()
```

## Histograms

מידע!

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115,135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

צרו היסטוגרמה מהמידע

```python
plt.hist(blood_sugar_men)
plt.show()
```

אנו יכולים להגדיר בעצמנו כמה בינים יהיו בגרף שלנו - נעשה זאת באמצעות הגדרת הפרמטר "bins = n" וכך יהיו לנו n בינים בטווח המידע שלנו.

נוכל גם ליצור רווחים בין הבינים באמצעות הפרמטר rwidth

וכך נחשב את האורך באופן יחסי לגודל הבינים

```python
plt.hist(blood_sugar_men, bins=20, rwidth=0.90)
plt.show()
```

נוכל להכניס לפרמטר bin גם רשימה של ערכים שיהוו את טווח הבינים שלנו.

נסו זאת בעצמכן

```python
plt.hist(blood_sugar_men, bins=range(75,155,20), rwidth=0.95)
plt.show()
```

כנסו לדוקומנטציה וחפשו מאפיינים תכונות של הפונקציה hist

ואם נרצה שהעמודות יהיו מופרדות?

*לא לשכוח כותרות ותוויות


```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=range(75,155,20), rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'])
plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.legend()
plt.show()
```

ועכשיו נהפוך את הגרף לאופקי

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=range(75,155,20), rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'], orientation='horizontal')
plt.xlabel('Sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')
plt.legend()
plt.show()
```

## Subplots
כעת, נלמד כיצד להציג מספר גרפים באותה תבנית 

קראו על הפונקציה subplots

צרו שני תרשימי קו אחד מעל השני והציגו אותם בתבנית אחת. השתמשו במידע על הטמפרטורה (max_t, min_t)

הציגו את הגרפים כך שציר הX שלהם יהיה משותף

```python
fig, axs = plt.subplots(2,sharex = True)
fig.suptitle('Temperatures')
axs[0].plot(days, max_t)
axs[1].plot(days, min_t)
plt.show()
```

כעת צרו שלושה תרשימי קו באותה תבנית,הוסיפו להם כותרות וצבעים שונים (השתמשו גם ב avg_t)

```python
fig, axs = plt.subplots(3,constrained_layout = True, sharex = True)
axs[0].plot(days, max_t,color = 'r')
axs[0].set_title("Maximum Temperature")
axs[1].plot(days, min_t,color = 'b')
axs[1].set_title("Minimum Temperature")
axs[2].plot(days, avg_t,color = 'orange')
axs[2].set_title("Average Temperature")
plt.show()
```

כעת הציגו שתי היסטוגרמות אחת ליד השנייה - השתמשו במידע על רמת הסוכר בדם אצל גברים ואצל נשים

```python
fig, axs = plt.subplots(1, 2)
fig.suptitle('Histograms')
axs[0].hist(blood_sugar_men)
axs[1].hist(blood_sugar_women)
plt.show()
```

התקבל מידע על טמפרטורות בישראל:

```python
israel_t = [52,44, 47, 43,40,47,49]
```

הציגו את המידע בתבנית לאורך (המידע על טמפרטורות), השתמשו בלולאה!

```python
temp = [max_t,min_t,avg_t,israel_t]
elements_ln = len(temp)
fig, axs = plt.subplots(elements_ln, sharex = True)
fig.suptitle('Temperatures')
for i in range(elements_ln):
    axs[i].plot(days, temp[i])
plt.show()
```

הגיע מידע נוסף על טמפרטורות מארצות הברית:

```python
usa_t = [40,43, 44, 44,42,44,39]
```

הוסיפו את המידע והציגו אותו,בצורה דינאמית(בלולאה)

```python
temp.append(usa_t)
elements_ln = len(temp)
fig, axs = plt.subplots(elements_ln)
fig.suptitle('Temperatures')
fig.subplots_adjust(hspace=0.5)  
for i in range(elements_ln):
    axs[i].plot(days, temp[i])
plt.show()
```

## ייצוא גרף לקובץ
אנו יכולים לשמור גרף כקובץ, קראו באינטרנט כיצד ושמרו את הגרף שיצרתם כקובץ עם נתיב מפורט

```python
fig, axs = plt.subplots(elements_ln)
fig.suptitle('Temperatures')
for i in range(elements_ln):
    axs[i].plot(days, temp[i])
plt.savefig(r'TempGraph.pdf', bbox_inches='tight', pad_inches=0.1, transparent=True)
```
