```python
#data
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)
data = np.column_stack((x, y))

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

#styling the figure
fig.suptitle("Subplots Example")

#plotting the left scatterplot
axs[0].scatter(x=x, y=y, marker='o', c='r', edgecolor='b')

#styling the first axle
axs[0].set_title('Scatter: $x$ versus $y$')
axs[0].set_xlabel('$x$')
axs[0].set_ylabel('$y$')

#plotting the right histogram
axs[1].hist(data, bins = 15, label=('x', 'y'))

#styling the second axle
axs[1].legend(loc = (0.9, 0.9))
axs[1].set_title('Frequencies of $x$ and $y$')
axs[1].yaxis.tick_right()
```
